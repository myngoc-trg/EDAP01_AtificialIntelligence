#### Discussion

This lab provided a concrete and instructive comparision of uninformed and informed search stategies through their application to different mazes. For maze 5, although several algorithms ultimately found the same optimal solution, their performance in terms of expanded nodes, memmory usage and search behavior differed significantly. While maze 5 provides a clear snapshot of strengths and weakneses, the result from maze 1-4 shows that these behaviors are not incidental but strutural, driven by how each algorithm explores the search space. These differences show the fundamental trade-offs between optimality, efficiency and resource consumption in search algorithms.

Breadth-First Search (BFS) demonstrated highly consistent behavior across all mazes. In every case, BFS found an optimal path, as evidence by the close correspondence between path cost and maximum tree depth. Uniform-Cost Search (UCS) behaved almost identically to BFS across all mazes. BFS and UCS both produced an optimal path with a cost of 51. This outcome is expected because all step costs in the maze are uniform, making UCS functionally equivalent to BFS in this environment. However, both algorithms expanded a large number of nodes (576) and required substantial memory to store the frontier and expanded sets. Due to them being uninformed searches, they expand nodes uniformly in all directions without regard to the heuristic or distance to the goal, which leads to unnecessary computation in large or open ares of the maze. While BFS and UCS achieve optimality when actions cost are all the same, this comes at a high computational cost.

Depth-First Search (DFS) shows the most volatile performance across the mazes. In some cases, such as maze 3, DFS managed to find an optimal path, but this outcome was largely coincidental rather than systematic. In most mazes, DFS produced paths with significantly higher costs and extremely large maximum depths, sometimes exceeding 4 times the optimal solution depth. DFS pursues a single path as deeply as possible before backtracking, which makes it vunerable to getting trapped in long corrdidors or dead ends. This behavior is especially visible in mazes where the goal is not located near the deepest nodes of the search tree, such as maze 1,2,4 and 5, which causes DFS to spend a large amount of time exploring deep, irrelevant branches before eventually encountering the goal. DFS performed the worst overall in maze 5. Although it expanded fewer nodes than BFS and UCS, it produces a significantly longer path with a cost of 155 and reached a maximum depth of 198. DFS returns the first goal found, regardless of depth or cost. Therefore, the resulting solution was highly inefficient, demonstrating that minimizing the number of expanded nodes does not necessarily correlate with finding a good solution. DFS's lack of optimality make it unsuitable for problems where solution quality is important.

Greedy Best-First Search (GBFS) consistently demonstrated the lowest number of expanded nodes and the smallest memory across all mazes. By relying solely on the heuristic being the Manhattan distance to guide the search toward the goal. Inseveral mazes, such maze 3 and maze 5, GBFS even managed to find optimal solutions with dramatically fewer expansions. In maze 5, It found the optimal path with a cost of 51 while expanding only 156 nodes and using the least amount of memory among all algorithms.  The strong performance of GBFS in maze 5 suggests that the heuristic aligned well with the maze's structure, allowing the algorithm to move efficiently toward the target. However, this success should be interpreted cautiously. In mazes 1,2, and 3, GBFS produced paths with higher costs than the optimal solutions. This is a limitation of GBFS. Because it ignores the accumulated path costs, it can be misled by heuristics into choosing locally attractive but globally supoptimal routes. GBFS does not garantee optimal solutions in general and can perform poorly in environments where the heuristic is misleading or where significant detours are required.

A* search gave a balanced compromise between efficiency and reliability across all mazes. In every case, A* found an optimal path while expanding significantly fewer nodes than BFS and UCS. Although it did not match GBFS’s efficiency in terms of node expansion, it consistently required less memory and avoided the severe inefficiencies seen in DFS.Like BFS and UCS in maze 5, A* found the optimal path with a cost of 51, but it expanded significantly fewer nodes (412) and required less memory. By conbining the actual path cost with the heuristic estimate, A* was able to focus the search algorithm toward promising regions while still maintaining optimality garantees. Although it was not as fast as GBFS in this maze, A* provides more consistent performance across different problem structures, making it a robust and generally preferable choice for pathfinding tasks.

Overall, this lab demonstrates that no single search algorithm is universally optimal. Uninformed methods such as BFS and UC provide strong guarantees but suffer from high time and memory costs. DFS is fast in terms of node expansion but unreliable and often suboptimal. Heuristic-based methods darmatically reduce search effort, with GBFS offering impressive efficiency when the heuristic is well suitéd to the problem, and A* providing the best balance between optimality and efficiency. Maze 5 clearly illustrates how heuristic quality and maze structure play a crucial role in determining search performance and emphasizes the importance of selecting an appropriate algorithm based on problem contraints and desired guarantees.
